# Line Plot

```python
# Create array as x
import numpy as np
from matplotlib import pyplot as plt

x = np.arange(1,11)
x
```

```python
# Create y
y = 2*x
y
```

```python
# Create line plot
# using plot
plt.plot(x,y)
plt.show()
```

```python
# Adding title and labels
plt.plot(x,y)
plt.title("Line Plot")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.show()
```

```python
# changing line aesthetics
plt.plot(x,y,color = 'g', linestyle = ':', linewidth = 2)
plt.show()
```

```python
# Adding two line same plot

x = np.arange(1,11)
y1 = 2*x
y2 = 3*x

plt.plot(x,y1, color='b',linestyle='-',linewidth=2)
plt.plot(x,y2, color='r',linestyle=':',linewidth=3)
plt.title('Line Plot')
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.grid(True)  # create grid
plt.show()
```

```python
# Adding sub plot

x = np.arange(1,11)
y1 = x*2
y2 = x*3

plt.subplot(1,2,1)
plt.plot(x,y1, color = 'g', linestyle = ':', linewidth = 2)
plt.subplot(1,2,2)
plt.plot(x,y2, color = 'b', linestyle = ':', linewidth = 2)

plt.show()
```

```python
from matplotlib import pyplot as plt
student = {"Bob":87,"Matt":56,"Sam":27} #create dict

names = list(student.keys())  #create variable from dict
values = list(student.values())

plt.bar(names,values) # create bar plot
plt.title("Bar Plot")
plt.xlabel('Student')
plt.ylabel('Grades')
plt.grid(True)
plt.show()
```

```python
# horizontal bar plot
plt.barh(names,values, color='g') # create barh for horizontal plot
plt.title("Bar Plot")
plt.xlabel('Grades')
plt.ylabel('Student')
plt.grid(True)
plt.show()
```

# Scatter Plot

```python
# basic scatter plot
x = [10,20,30,40,50,60,70,80,90]
a = [3,2,5,4,2,9,8,3,7]

plt.scatter(x,a)
plt.grid(True)
plt.show
```

```python
# aesthetics scatter plot
x = [10,20,30,40,50,60,70,80,90]
a = [3,2,5,4,2,9,8,3,7]

plt.scatter(x,a,marker='*',c='orange',s=100)
plt.show
```

```python
# double scatter plot
# aesthetics scatter plot
x = [10,20,30,40,50,60,70,80,90]
a = [3,2,5,4,2,9,8,3,7]
b = [7,4,6,3,8,4,7,4,9]

plt.scatter(x,a,marker='*',c='orange',s=100)
plt.scatter(x,b,marker='o',c='red',s=100)
plt.show
```

```python
# scatter sub plot
# aesthetics scatter plot
x = [10,20,30,40,50,60,70,80,90]
a = [3,2,5,4,2,9,8,3,7]
b = [7,4,6,3,8,4,7,4,9]

plt.subplot(1,2,1)
plt.scatter(x,a,marker='*',c='orange',s=100)

plt.subplot(1,2,2)
plt.scatter(x,b,marker='o',c='red',s=100)

plt.show
```

# Histogram

```python
# Creating data
x = [6,1,3,3,3,4,4,5,5,5,7,7,7,7,2,2,8,8,8]

plt.hist(x)
plt.show
```

```python
# aesthetics histogram
x = [6,1,3,3,3,4,4,5,5,5,7,7,7,7,2,2,8,8,8]

plt.hist(x,color = 'g',bins=5)
plt.show
```

# Box Plot

```python
# box plot = interval data
one = [1,2,3,4,5,6,7,8,9]
two = [1,2,3,4,5,4,3,2,1]
three = [5,7,9,6,5,8,9,5,7]

x = list([one,two,three])
plt.boxplot(x)
plt.show
```

# Violin Plot

```python
# violin plot
one = [1,2,3,4,5,6,7,8,9]
two = [1,2,3,4,5,4,3,2,1]
three = [5,7,9,6,5,8,9,5,7]

x = list([one,two,three])
plt.violinplot(x,showmedians=True)
plt.show()
```

# Pie Chart

```python
# Basic pie chart
fruit = ['mango','apple','grapes','durian']
quantity = [55,25,60,100]

plt.pie(quantity,labels=fruit)
plt.show()
```

```python
# Aesthetic pie chart
# color and percent
fruit = ['mango','apple','grapes','durian']
quantity = [55,25,60,100]

plt.pie(quantity,labels=fruit, autopct='%0.1f%%',colors=['green','red','purple','yellow'])
plt.show()
```

# Doughnut Chart

```python
# Basic doughnut chart
fruit = ['mango','apple','grapes','durian']
quantity = [55,25,60,100]

plt.pie(quantity,labels=fruit,radius=1)
plt.pie([1],colors=['w'],radius=0.5)  #adding white circle at core
plt.show()
```
