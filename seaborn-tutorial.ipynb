# Seaborn

# Line Plot

```python
import seaborn as sns
```

```python
# load dataset fmri from seaborn library
df = sns.load_dataset('fmri')
df.head()
```

```python
# Create line plot
from matplotlib import pyplot as plt

sns.lineplot(x='timepoint',y='signal',data=df)
plt.show()
```

```python
# Grouping data with hue

sns.lineplot(x='timepoint',y='signal',data=df,hue='event')
plt.show()
```

```python
# Adding style

sns.lineplot(x='timepoint', y='signal', data=df, hue='event', style='event')
plt.show()
```

```python
# Adding marker

sns.lineplot(x='timepoint', y='signal', data=df, hue='event', style='event', markers=True)
plt.show()
```

# Count Plot

```python
# load dataset

import pandas as pd
from matplotlib import pyplot as plt

sns.set(style="whitegrid")
df = pd.read_csv('data.csv')
df.head()
```

```python
# Create count plot
sns.countplot(x=df['Duration'],hue=df['Duration'])
plt.show()
```

# Bar Plot

```python
# load dataset

import pandas as pd

sns.set(style="whitegrid")
df = pd.read_csv('data.csv')
df.head()
```

```python
# Create bar plot from dataset column
sns.barplot(x='Duration',y='Pulse',data=df)
plt.show()
```

```python
# Color bar plot from dataset column
sns.barplot(x='Duration',y='Pulse',data=df, palette="rocket")     #pick paltte(Blues_d/rocket/vlag) OR color
plt.show()
```

```python
# Adding hue color to bar plot

sns.barplot(x="Duration",y="Calories",data=df, hue="Pulse") # hue from duration column
plt.show()
```

# Joint Plot

```python
# load dataset
import pandas as pd

x = pd.read_csv("data.csv")
x.head()
```

```python
# Create jointplot 

sns.jointplot(x='Duration',y='Pulse',data=x,color='olive',kind='reg')   # reg = regression line
plt.show()
```

# Box plot

```python
# Create box plot
sns.boxplot(x='Duration',y='Pulse',data=x,linewidth=2,hue='Duration')
plt.show()
```
